## Mean, Median, Mode

Given the data:

$$ X = {x_1, x_2, x_3, ..., x_i}$$

We  have:

- **mean**: $ \frac{(Î£x)}{n}$

- **median**: $x_i$ represents the $i-th$ value in the **sorted** dataset, and $n$ represents the **total number of values** in the dataset. $$
\begin{cases}
x_{\frac{n+1}{2}} & \text{if } n \text{ is odd} \\
\frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2} & \text{if } n \text{ is even}
\end{cases}$$

- **mode**: The most frequently occurring value(s) in the dataset

### The Data

```python
import numpy as np
from scipy import stats

# Setting random seed for reproducibility
np.random.seed(0)

# Generating random data
data = np.random.randint(10, size=(10))
print(f'Data: {data}\n')
```

### Visualization

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=range(min(data) - 1, max(data) + 2), edgecolor='black', align='left')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Data')
plt.xticks(range(min(data), max(data) + 1))
plt.show()
```

### The Mean

```python
# Mean - method 1 - using numpy
Mean = np.mean(data)
print(f'Method 1 - using numpy: The mean (average) of the data is: {Mean}!\n')

# Mean - method 2 - custom function
def mean(arr):
  # Sum(X)
  Sum = np.sum(arr)
  # number of values in X
  n = len(arr)
  return Sum/n

Mean = mean(data)
print(f'Method 2 - custom function: The mean (average) of the data is: {Mean}!\n')
```

### The Median

```python
# Median - method 1 - using numpy
Median = np.median(data)
print(f'Method 1 - using numpy: The median (middle value) of the data is: {Median}!\n')

# Median - method 2 - custom function
def median(arr):
  # Sorting the X in ascending order
  arr.sort()
  # Finding the number of elements in X
  n = len(arr)
  # if arr has odd len (n): X[(n+1)/2]
  if len(arr) % 2 != 0:
    return arr[n // 2]
  # if arr has even len (n): (X[(n)/2]+X[(n-1)/2])/2
  # Note: indices start from 0. (e.g. [0,1,2,3] so 1,2 are middle)
  elif len(arr) % 2 == 0:
    return (arr[n // 2 - 1] + arr[n // 2]) / 2

Median = median(data)
print(f'Method 2 - custom function: The median (middle value) of the data is: {Median}!\n')
```

### The Mode

```python
# Mode - method 1 - using scipy
Mode = stats.mode(data)
print(f'Method 1 - using scipy: The mode (most repeated) of the data is: {Mode[0]}, which is repeated {Mode[1]} times!\n')

# Mode - method 2 - custom function
def mode(arr):
    # Create a dictionary to store the frequency of each element
    frequency = {}
    # Iterate through the array and count the frequency of each element
    for num in arr:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    # Find the mode(s) by finding the maximum frequency
    max_frequency = max(frequency.values())
    modes = [key for key, value in frequency.items() if value == max_frequency]
    return modes[0], max_frequency

Mode = mode(data)
print(f'Method 2 - custom function: The mode (most repeated) of the data is: {Mode[0]}, which is repeated {Mode[1]} times!\n')
```

### Range, Variance, Standard Deviation

Given the data:

$$ X = {x_1, x_2, x_3, ..., x_i}$$

We  have:

- **Range**: $Max(X) - Min(X)$

- **Variance**: $x_i$ = data points, $Î¼$ = mean of the data, $N$ = number of data points. $$\begin{cases}
\text{population variance = } \frac{1}{N} ðº (x_i - Î¼)^2 \\
\text{sample variance = }  \frac{1}{N-1} ðº (x_i - Î¼)^2

- **Standard deviation**:$$\begin{cases}
\text{population variance = } \sqrt{\frac{1}{N} ðº (x_i - Î¼)^2} \\
\text{sample variance = }  \sqrt{\frac{1}{N-1} ðº (x_i - Î¼)^2}

### Variance


```python
# Variance - method 1 - using numpy
Variance = np.var(data)
print(f'Method 1 - using numpy: The variance of the data is: {Variance}!\n')

# Range - method 2 - custom function
def variance(arr):
  # Step 1: Calculate the mean of the array
  mean = np.mean(arr)
  # Step 2: Calculate the squared differences between each element and the mean
  squared_diff = [(num - mean)**2 for num in arr]
  # Step 3: Calculate the sum of squared differences
  sum_squared_diff = np.sum(squared_diff)
  # Step 4: Calculate the variance
  variance = (1/len(arr))*sum_squared_diff
  return variance

Variance = variance(data)
print(f'Method 2 - custom function: The variance of the data is: {Variance}!\n')
```

```python
# Calculate variance
variance = np.var(data)

# Calculate standard deviation
std_dev = np.sqrt(variance)

# Plotting Variance
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Variance Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=1, label='Mean')

# Plot lines for different multiples of standard deviation
for i in range(1, 4):
    plt.axvline(np.mean(data) + i * std_dev, color='green', linestyle='dashed', linewidth=1, label=f'Mean + {i}$\sigma$')
    plt.axvline(np.mean(data) - i * std_dev, color='green', linestyle='dashed', linewidth=1, label=f'Mean - {i}$\sigma$')

plt.legend()
plt.grid(True)
plt.show()
```

### Standard deviation

```python
# Standard deviation - method 1 - using numpy
Standard_deviation = np.std(data)
print(f'Method 1 - using numpy: The Standard deviation of the data is: {Standard_deviation}!\n')

# Standard deviation - method 2 - custom function
def Standarddeviation(arr):
    # Step 1: Calculate the mean of the array
    mean = np.mean(arr)
    # Step 2: Calculate the squared differences between each element and the mean
    squared_diff = [(num - mean)**2 for num in arr]
    # Step 3: Calculate the sum of squared differences
    sum_squared_diff = np.sum(squared_diff)
    # Step 4: Calculate the variance
    variance = (1/len(arr)) * sum_squared_diff
    # Step 5: Calculate the standard deviation by taking the square root of the variance
    std_deviation = np.sqrt(variance)
    return std_deviation

Standard_deviation = Standarddeviation(data)
print(f'Method 2 - custom function: The Standard deviation of the data is: {Standard_deviation}!\n')
```

## Problem Statement:

A beverage company claims that their new energy drink increases the average reaction time of individuals by at least 10 milliseconds. To test this claim, a random sample of 30 individuals is selected, and their reaction times are recorded before and after consuming the energy drink. Assume the population standard deviation of reaction times is 8 milliseconds. Use a significance level of 0.05.

### Hypotheses:

**Null Hypothesis (H0)**: The average reaction time before and after consuming the energy drink is the same or decreases. ($Î¼_{difference} <= 0$)

**Alternative Hypothesis (H1)**: The average reaction time after consuming the energy drink is greater than the average reaction time before consuming it. ($Î¼_{difference} > 0$)

- **Type I Error**: Rejecting the null hypothesis when it is actually true.
- **Type II Error**: Failing to reject the null hypothesis when the alternative hypothesis is true.

```python
import numpy as np
from scipy import stats

# Sample data (reaction times before and after consuming the energy drink)
reaction_times_before = np.array([220, 210, 230, 215, 225, 205, 235, 240, 215, 225, 230, 220, 210, 225, 235, 220, 225, 215, 230, 240, 215, 210, 220, 225, 230, 235, 240, 215, 225, 235])
reaction_times_after = np.array([210, 200, 220, 205, 215, 195, 225, 230, 205, 215, 220, 210, 200, 215, 225, 210, 215, 205, 220, 230, 205, 200, 210, 215, 220, 225, 230, 235, 210, 220])

# Calculate the differences in reaction times
differences = reaction_times_after - reaction_times_before

# Sample size
n = len(differences)

# Population standard deviation
sigma = 8

# Hypothesized difference
hypothesized_difference = 10

# Significance level
alpha = 0.05

# Calculate the test statistic
t_statistic = (differences.mean() - hypothesized_difference) / (sigma / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Calculate the critical value from the t-table
critical_value = stats.t.ppf(1 - alpha, df)

# Calculate the lower tail p-value from the t-table
p_value = 1 - stats.t.cdf(t_statistic, df)

print("Test Statistic:", t_statistic)
print("Critical Value from t-table:", critical_value)
print("Lower tail p-value from t-table:", p_value)

# Analyze the results
if t_statistic > critical_value:
    print("Reject the null hypothesis (H0). There is sufficient evidence to conclude that the energy drink increases the average reaction time (supporting H1).")
else:
    print("Fail to reject the null hypothesis (H0). There is not enough evidence to conclude that the energy drink increases the average reaction time.")
```

#### Analysis:

- The test statistic obtained is -13.24, which represents how many standard errors the sample mean is away from the hypothesized mean under the null hypothesis. Since the test statistic is far below the critical value of 1.70 and the lower tail p-value is close to 1 (0.9999999999999597), we fail to reject the null hypothesis.

- Therefore, we do not have enough evidence to conclude that the energy drink increases the average reaction time. This suggests that, based on the sample data, there is no significant difference in reaction times before and after consuming the energy drink.

## Normal distribution

**Z-Score Calculation**:

The Z-score (also known as standard score) measures how many standard deviations a data point is from the mean of the distribution. It is calculated using the formula:
$$Z = \frac{(X - Î¼)}{Ï}$$

Where:

$X$ is the value (in our case, the threshold height),
$Î¼$ is the mean of the distribution (in our case, the mean height),
$Ï$ is the standard deviation of the distribution (in our case, the standard deviation of heights).

**Probability Calculation**:
Once we have calculated the Z-score, we can use it to find the probability of obtaining a value equal to or greater than the threshold value. This is done by finding the area under the normal distribution curve to the right of the Z-score.

The cumulative distribution function (CDF) of the standard normal distribution, denoted by $Î¦(z)$, gives the probability that a random variable takes a value less than or equal to a given value. In our case, the given value is the Z-score $z$. Therefore, the CDF gives us the probability of obtaining a value less than or equal to the Z-score.

Since we are interested in finding the probability of obtaining a value greater than the threshold value (not less than or equal to), we need to subtract the probability obtained from the CDF from 1.

So, the formula to find the probability of $X$ being greater than $x$ (where $x$ is the threshold value) is:

$$P(X > x) = 1 - Î¦(z)$$

Where:

- $Î¦(z)$ is the cumulative distribution function of the standard normal distribution.
- $z$ is the Z-score calculated earlier.

```python
import scipy.stats as stats

# Hypothesized parameters
mean_height = 175  # Mean height in centimeters
std_deviation = 7  # Standard deviation in centimeters

# Threshold height for analysis
threshold_height = 185  # Height above which we want to calculate the probability

# Calculate the z-score corresponding to the threshold height
z_score = (threshold_height - mean_height) / std_deviation

# Calculate the probability of selecting an adult male taller than the threshold height
probability = 1 - stats.norm.cdf(z_score)

# Convert probability to percentage
percentage = probability * 100

print(f"The probability of randomly selecting an adult male taller than {threshold_height} centimeters is approximately {percentage:.2f}%")
```

## Chi-square:

**Chi-square $(ÏÂ²)$** is a statistical test used to determine whether there is a significant association between categorical variables. It is often used to analyze data where observations are sorted into categories to understand if there is a difference between the expected frequencies and the observed frequencies in one or more categories.

The formula for the Chi-square test statistic is:

$$ÏÂ² = Î£ ((O - E)Â² / E)$$

Where:

- $ÏÂ²$ is the Chi-square test statistic.
- $O$ is the observed frequency.
- $E$ is the expected frequency.
- $Î£$ represents the sum across all categories.

**Hypothesis**:

Let's consider an example scenario where we want to test whether there is a significant association between gender and preference for a certain type of music. Our hypothesis would be:

|       | Pop | Rock | Classical |
| ----| ---|---|---|
| Male  |  30  |  20   |    10 |
| Female | 15  |  25   |    20 |

- **Null Hypothesis (H0)**: There is no significant association between gender and music preference.
- **Alternative Hypothesis (H1)**: There is a significant association between gender and music preference.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example data: Gender vs. Music Preference
# Create a contingency table
#        Pop  Rock  Classical
# Male    30    20       10
# Female  15    25       20

observed = np.array([[30, 20, 10],
                     [15, 25, 20]])

# Perform Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(observed)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

**Analysis**

The Chi-square test conducted on the relationship between gender and music preference yielded a Chi-square statistic of 8.8889 with a corresponding p-value of 0.0117, both indicating a significant association between the variables. If $p-value < 0.05$, there is a strong dependency between two variables, whereas if $p-value > 0.05$, there is no dependency between the variables.




With two degrees of freedom, the analysis suggests that gender influences music preference. The expected frequencies, computed at 22.5 for each genre preference among males and females, underscore the deviation from what would be anticipated under the null hypothesis of no association. Thus, the rejection of the null hypothesis leads to the conclusion that gender plays a discernible role in determining music preference among individuals.

## ANOVA

Variance analysis evaluates whether the means of multiple populations are equal or not. ANOVAs gauge the significance of one or more factors by contrasting the means of the response variable across various levels of these factors. The null hypothesis posits that all population means are identical, whereas the alternative hypothesis suggests that at least one mean differs.

**Hypothesis**

A fertilizer company developed three new types of universal fertilizers after research that can be utilized to grow any type of crop. In order to find out whether all three have a similar crop yield, they randomly chose six crop types in the study. In accordance with the randomized block design, each crop type will be tested with all three types of fertilizer separately. The following table represents the yield in g/m2. At the 0.05 level of significance, test whether the mean yields for the three new types of fertilizers are all equal:

|Fertilizer 1| Fertilizer 2| Fertilizer 3 |
|---|---|---|
|62| 54| 48 |
|62| 56| 62 |
|90| 58| 92 |
|42| 36| 96 |
|84| 72| 92 |
|64| 34| 80|

```python
import pandas as pd
from scipy import stats

d = {'Fertilizer 1 ': [62, 62, 90, 42, 84, 64], 'Fertilizer 2': [54, 56, 58, 36, 72, 34], 'Fertilizer 3': [48, 62, 92, 96, 92, 80]}
fertilizers = pd.DataFrame(data=d)

one_way_anova = stats.f_oneway(fertilizers["Fertilizer 1 "], fertilizers["Fertilizer 2"], fertilizers["Fertilizer 3"])

print ("Statistic :", round(one_way_anova[0],2),", p-value:",round(one_way_anova[1],3))
```

**Analysis**

The obtained p-value is less than 0.05, indicating that we can reject the null hypothesis stating that the mean crop yields of the fertilizers are equal. This suggests that fertilizers have a significant impact on crop yields.

## **Confusion matrix**:

This is the matrix of the actual versus the predicted. This concept is better explained with the example of cancer prediction using the model:

||Predicted: yes|	Predicted: no|
|--|---|--|
|Actual: yes|	True Positive (TP)	|False Negative (FN)|
|Actual: no	|False Positive (FP)	|True Negative (TN)|

Certainly, here's a rewritten version:

- **True Positives (TPs)** occur when we correctly identify cases where the patient has the disease.
- **True Negatives (TNs)** are instances where we accurately determine that the patient does not have the disease.
- **False Positives (FPs)** happen when we incorrectly diagnose the disease in a patient who does not have it; these are also known as Type I errors.
- **False Negatives (FNs)** arise when we miss identifying the disease in a patient who actually has it; these are also known as Type II errors.
- **Precision (P)** When yes is predicted, how often is it correct?
$$\frac{TP}{TP+FP}$$
- **Recall (R)/sensitivity/true positive rate**: Among the actual yeses, what fraction was predicted as yes?

$$\frac{TP}{TP+FN}$$

- **F1 score (F1)**: the F1 score is a way to combine the precision and recall of a classification model into a single metric. It provides a balance between these two measures.

**Example**

Imagine you have a medical test to detect a disease. Precision tells you how accurate the test is in identifying the disease when it says a person has it. Recall tells you how good the test is at finding all the people who actually have the disease.

The F1 score is like looking at both precision and recall together. It's calculated using the harmonic mean of precision and recall. If both precision and recall are high, then the F1 score will also be high.

For example, let's say you have a medical test with a precision of 80% and a recall of 70%. To calculate the F1 score, you'd use this formula:

$F1 Score (F1) = \frac{2(P \times R)}{P+R}$

Substituting the values:

$F1 Score (F1) = \frac{2(0.8 \times 0.7)}{0.8 + 0.7}$

$F1 Score â 0.746$

So, in this example, the F1 score would be approximately 0.746, indicating a good balance between precision and recall.

- **Specificity**: Among the actual nos, what fraction was predicted as no? Also equivalent to 1- false positive rate:

$$\frac{TN}{TN+FP}$$

- **Area under curve (ROC)**: Receiver operating characteristic curve is used to plot between true positive rate (TPR) and false positive rate (FPR), also known as a sensitivity and 1- specificity graph. The area under the ROC curve (AUC-ROC) is a scalar value between 0 and 1. A higher AUC-ROC value indicates better overall performance of the classifier in distinguishing between the positive and negative classes. An AUC-ROC of 1 represents a perfect classifier, while an AUC-ROC of 0.5 suggests a classifier with no discriminatory power (essentially random guessing). In summary, the AUC-ROC is a metric used to quantify the performance of a binary classification model, with higher values indicating better discrimination between positive and negative cases.

## R-squared (Coefficient of Determination)

R-squared, denoted as \( R^2 \), is a measure of the percentage of the response variable variation that is explained by a model. It quantifies how well the model minimizes error compared with just utilizing the mean as an estimate.

In mathematical terms, it is defined as:

$$
R^2 = 1 - \frac{SS_{\text{residual}}}{SS_{\text{total}}}
$$

Where:
- $SS_{\text{residual}}$ is the sum of squares of the residuals (the differences between the observed and predicted values). It's calculated as: $SS_{\text{residual}} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
  Where $n$ is the number of observations, $y_i$ is the observed value, and $\hat{y}_i$ is the predicted value.

- $SS_{\text{total}}$ is the total sum of squares, representing the total variance of the response variable around its mean. It's calculated as: $SS_{\text{total}} = \sum_{i=1}^{n} (y_i - \bar{y})^2$ Where $\bar{y}$ is the mean of the observed values $y_i$.

The R-squared value ranges from 0 to 1. A value closer to 1 indicates that a larger proportion of the variance in the response variable is explained by the model. Conversely, a value closer to 0 suggests that the model does not explain much of the variance.

It's important to note that in some extreme cases, R-squared can have a value less than zero, indicating that the predicted values from the model perform worse than just taking the simple mean as a prediction for all the observations.
